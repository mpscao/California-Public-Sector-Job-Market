```python
df1 = pd.read_csv('10122023_USA Staffing_Cleaned (2022q2)_opm_included.csv')
```

```python
df1
```

```python
# find all nursing variables, which contains code number 00 in grade
df2 = df1[df1['grade'].str.contains('00')==True]
df2
```

```python
# remove the appostrophe in job_series_number
df2['job_series_number'] = df2['job_series_number'].str.replace(r"'", '')
```

```python
df2
```

```python
# split the numbers after and before slash into two seperate values and make them into seperate rows
df3 = df2.assign(job_series_number=df2['job_series_number'].str.split('/')).explode('job_series_number', ignore_index=True)
df3
```

```python
df3['job_series_number'] = df3['job_series_number'].astype(int)
```

```python
df3
```

```python
new_data = df.append(df3)
```

```python
new_data
```

```python
jobs = pd.read_csv('jobs.csv')
```

```python
# split the excel file 
bh = jobs['Number'].apply(lambda x: x.split('-')).apply(pd.Series).astype(int).values[:,0]
bt = jobs['Number'].apply(lambda x: x.split('-')).apply(pd.Series).astype(int).values[:,1]
bins = pd.IntervalIndex.from_arrays(bh, bt, closed='both')

print(bins)
```

```python
# cut the bins to correspond each job number to the group it falls under and then use zip to give each job number its correct classification.
new_data['position_classification_plan'] = pd.cut(new_data['job_series_number'], bins = bins).map(dict(zip(bins, jobs['Group'])))
```

```python
new_data
```

```python
new_data = new_data.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
```
